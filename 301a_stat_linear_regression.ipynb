```python
import numpy as np
import matplotlib.pyplot as plt
```

### Equation for Linear Regression

In a linear regression model with one feature (simple linear regression), the goal is to find the best-fit line:

ùë¶ = ùë§ùëã + ùëè
Where:

X is the independent variable.
y is the dependent variable.
w is the slope (weight).
b is the intercept (bias).


Estimating the Parameters 
ùë§ and ùëè

Given 
ùëõ data points 

Slope (Weight) 
$$
w = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}

$$
Intercept (Bias) 
$$
b = \bar{y} - w \bar{x}
$$

```python
# Creating dataset

np.random.seed(42)
X = np.random.rand(100,1) * 10
print(X.shape)
print(X[:10])
```

```python
# y = 2x + 3 + noise

y = 2 * X + 3 + np.random.randn(100,1) * 2
print(y.shape)
print(y[:10])
```

```python
class StatLinearRegression():

    def __init__(self):
        self.w = None
        self.b = None

    def fit(self,X,y):
        # 1D array

        x_flat = X.flatten()
        y_flat = y.flatten()

        # mean
        x_mean = np.mean(x_flat)
        y_mean = np.mean(y_flat)

        # weights

        num = np.sum((x_flat - x_mean) * (y_flat - y_mean))
        den = np.sum((x_flat - x_mean)**2)
        self.w = num / den

        # intercept

        self.b = y_mean - self.w*x_mean

        print(f"Slope (w): {self.w:.4f}")
        print(f"Intercept (b): {self.b:.4f}")
        return (self.w,self.b)
    
    def predict(self,x):
        return self.w * x + self.b
```

```python
regressor = StatLinearRegression()
regressor.fit(X,y)
```

```python
regressor.predict(2)
```

```python
y_pred = regressor.predict(X)
```

```python
plt.scatter(X,y, label = 'Data points')
plt.plot(X, y_pred, color='r', label=f'Regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Statistical Linear Regression')
plt.legend()
plt.show()
```
